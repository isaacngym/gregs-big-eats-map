```python
# i opened the page, went to the videos tab on his page, scrolled till all videos loaded, then saved the page offline. 
# i'm guessing youtube dynamically loads content. it might not. but in any case beautifulsoup can't handle that.
with open("../data/Gregory Leow - YouTube.html", "r") as f:
    lines = f.readlines()
soup = BeautifulSoup("\n".join(lines), "lxml")
```

```python
# first find the grid renderer because the html of the page has a lot of other videos in it - so that the content is there 
# when you navigate to another tab on the same page, probably
grid_renderer = soup.find_all("ytd-grid-renderer")
assert len(grid_renderer) == 1
# then get all the video links
titles = grid_renderer[0].find_all("a", attrs = {"id":"video-title"})
assert len(titles) == 115 #at time of coding
```

```python
urls_to_scrape = set()
for title in titles:
    urls_to_scrape.add(title.attrs["href"])
```

```python
for url in urls_to_scrape:
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, "lxml")
    temp = soup.find("div", id = "action-panel-details")
    txt = temp.text
    # how now brown cow
    #TODO
    matching = re.findall("([A-Za-z0-9 ]+)([0-9]{6})([A-Za-z0-9 ]+)", txt) #find text around a postal code
    address = re.findall("(.+)(address)(.+)", txt, re.IGNORECASE)
    print(matching, address, url)
```
